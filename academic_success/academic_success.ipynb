```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import StandardScaler, OrdinalEncoder, OneHotEncoder, Normalizer
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import mutual_info_regression, mutual_info_classif
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.columns
```

```python
for i in df.columns:
    print("**"*10)
    print(i)
```

```python
df["Target"]
```

```python
myordenc = OrdinalEncoder()

df[["Target", ]] = pd.DataFrame(myordenc.fit_transform(df[["Target", ]]))
```

```python
df["Target"] = df["Target"].astype(np.int64)
```

```python
df.corr()
```

```python
# int_cols = [col for col in df.columns if df[col].dtype == np.int64][: -1]

mi_values = mutual_info_classif(df[df.columns[1 : -1]], df["Target"])

for i in range(len(df.columns[: -2])) :
    print(df.columns[i + 1], mi_values[i])
```

```python
X_cols = df.columns[1 : -1]
# X_cols = ["Course", "Admission grade", 
# "Tuition fees up to date", "Curricular units 1st sem (grade)", 
# "Curricular units 2nd sem (grade)"]

Y_cols = df.columns[-1]

X_train, X_test, Y_train, Y_test = train_test_split(df[X_cols], df[Y_cols], test_size = 0.1)
```

```python
from sklearn.svm import SVC
```

```python
model = make_pipeline(StandardScaler(), SVC())
model.fit(X_train, Y_train)

preds = model.predict(X_test)
accuracy_score(Y_test, preds)
```
